# Word Cloud Iphone

```python
import pandas as pd
from wordcloud import WordCloud
import matplotlib.pyplot as plt
#from scipy.misc import imread
#scipy.imageio.imread
from imageio import imread
dfIphone=pd.read_csv('C:/Users/pk/Desktop/projet/analyse/out_data/TweetIphone.csv')
dfIphonePositive=dfIphone[dfIphone.sentiment=='Positive'].iloc[:3,:]
dfIphoneNegative=dfIphone[dfIphone.sentiment=='Negative'].iloc[:3,:]
text=' '.join(dfIphoneNegative['CommentAferPreproc'])
maskTwitter=imread('iphoneLogo.png')
wordcloud_ip = WordCloud(
                      #font_path='CabinSketch-Bold.ttf',
                      background_color='white',
                      width=1800,
                      height=1400,
                      random_state = 2016,
                      #mask=maskTwitter
                      colormap='Reds'
                     ).generate(text)
wordcloud_ip.to_file('IphoneNegative.png')
plt.title("Tweets with word 'narendramodi'")
plt.imshow(wordcloud_ip)
```

# Word Cloud Sumsang

```python
import pandas as pd
from wordcloud import WordCloud
import matplotlib.pyplot as plt
#from scipy.misc import imread
#scipy.imageio.imread
from imageio import imread
dfSum=pd.read_csv('C:/Users/pk/Desktop/projet/analyse/out_data/TweetAfterProcess.csv')
dfSumPositive=dfSum[dfSum.sentiment=='Positive'].iloc[:20,:]
dfSumNegative=dfSum[dfSum.sentiment=='Negative'].iloc[:20,:]
textNegative=' '.join(dfSumNegative['CommentAferPreproc'])
textPositive=' '.join(dfSumPositive['CommentAferPreproc'])

wordcloud_ip =  WordCloud(background_color='black', width=800, height=500,colormap='Greens').generate(textPositive)
wordcloud_ip.to_file('SumPositive.png')
plt.title("Tweets with word 'narendramodi'")
plt.imshow(wordcloud_ip)
```
